# IMPORT DATA AND CLEAN 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib import style
style.use('ggplot')
```

```python
df = pd.read_csv('kc_house_data.csv')
X = np.array(df['sqft_living'])
Y = np.array(df['price'])

plt.scatter(X,Y)
```

```python
df.head()
```

```python
df = df.drop(['yr_renovated','lat','long'], axis=1)
```

```python
df.info()
```

Locating and replacing null values. 

```python
df.loc[df['waterfront'] == 1]
"""Replacing null values in the waterfront and views columns with zeros."""
df['waterfront'] =df['waterfront'].replace(np.nan, 0)
df['view'] = df['view'].replace(np.nan,0)
```

```python
df['zipcode'] = df['zipcode'].astype(str)
'''Changed zipcode to string, so that it doesn't skew data'''
```

```python
df.info()
```

```python

df['sqft_basement'] = df['sqft_basement'].replace("?",0)
df.head()
```

```python
sns.set(style="white")

# Create a covariance matrix
corr = df.corr()

# Generate a mask the size of our covariance matrix
mask = np.zeros_like(corr, dtype=bool)
mask[np.triu_indices_from(mask)] = True

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11,9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(220,10,as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr,mask=mask,cmap=cmap,vmax=.3,center=0,square=True, 
            linewidth=.5, cbar_kws={'shrink': .5})
```
